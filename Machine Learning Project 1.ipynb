```python
#08


# Handling outliers

# Using the Interquartile Range (IQR) method by calculating quartiles ,defining the threshold multiplier,

# then calculating the lower and upper bounds for potential outliers and creating a new dataframe without potential outliers

# (identifying and removing potential outliers)


mydf = mydf

# Identifying rows where any value in the row is an outlier based on the IQR method

# Calculate lower and upper bounds for outliers

Q1 = mydf.quantile(0.25)
Q3 = mydf.quantile(0.75)
IQR = Q3 - Q1
l_bound = Q1 - 1.5 * IQR # I tried putting in 3.0 but with no significant difference
u_bound = Q3 + 1.5 * IQR

# Aligning the indices of the bounds with the DataFrame

l_bound, u_bound = l_bound.align(u_bound, axis=0, copy=False)

# Identifying outliers

p_outliers = ((mydf < l_bound) | (mydf > u_bound)).any(axis=1)

# Printing the rows with outliers

mydf[p_outliers]
```

```python
# R-squared: 0.68

# R-squared (coefficient of determination) is a measure that indicates the proportion of variance in the dependent variable,which is explained by the independent variables in the model
# R-squared value of approximately 0.68 means that the model explains about 68% of the variance in the dependent variable suggesting that the model is catching a significant portion of the variability in the data

# Adjusted R-squared: 0.65
# Adjusted R-squared is a modified version of R-squared,which takes into account the number of independent variables in the model
# Also it penalizes the addition of unnecessary variables that don't contribute much to explaining the variance in the dependent variable.# so 
# adjusted R-squared value of approximately 0.65 indicates that this model adjusts for the number of variables while explaining a reasonable amount of variance in the dependent variable

# Mean Absolute Error: 3.68

# Mean Absolute Error (MAE) is a measure of the average absolute difference between the predicted values and the actual values
# MAE of approximately 3.65 indicates that on average, the model's predictions deviate from the actual values by about 3.65 units
# Coefficients:
# These are the coefficients of the independent variables in the regression model
# Each coefficient tells the change in the dependent variable for a one-unit change in the corresponding independent variable ath the same time holding other variables constant
# Positive coefficients indicate a positive relationship (increase in independent variable leads to an increase in the dependent variable), and negative coefficients indicate a negative relationship
# Here a positive coefficient of 5.02 for the variable suggests that a one-unit increase in that variable is associated with an increase of approximately 5.02 units in the dependent variable
# Intercept: 9.96

# The intercept is the value of the dependent variable when all independent variables are set to zero
# Giving a baseline value of the dependent variable when no independent variables are present there
# value of 10 might not have a direct intuitive interpretation on its own here as the Boston Housing dataset contains features which are not necessarily centered around the value zero
```

```python


# 20

# Applying crossvalidation to check the model performance

from sklearn.model_selection import cross_val_score

# Performing cross-validation

crossval_scores = cross_val_score(model1, X, y, cv=5, scoring='neg_mean_squared_error')

# Converting negative MSE scores to positive

crossval_scores = -crossval_scores

# Calculating mean and standard deviation of the cross-validation scores

mean_crossval_score = crossval_scores.mean()

std_crossval_score = crossval_scores.std()

#  Results

print("Cross-Validation Mean MSE:", mean_crossval_score)
print("Cross-Validation Standard Deviation:", std_crossval_score)
```
